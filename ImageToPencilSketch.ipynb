Original Image

```python
image = cv2.imread('image1.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,6))
plt.imshow(image)
plt.axis('off')
plt.title('ORIGINAL IMAGE')
plt.show()
```

Original to Grayscale

```python
grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,6))
plt.imshow(grayscale, cmap='gray')
plt.axis('off')
plt.title('GRAYSCALE IMAGE')
plt.show()
```

Invert Grayscale

```python
inverted = 255 - grayscale
plt.figure(figsize=(8,6))
plt.imshow(inverted,cmap='gray')
plt.axis('off')
plt.title('Inverted image')
plt.show()
```

Blurring Inverted Image

```python
blurred = cv2.GaussianBlur(inverted, (21, 21), 0)
plt.figure(figsize=(8,6))
plt.imshow(blurred,cmap='gray')
plt.axis('off')
plt.title('Blurred Inverted')
plt.show()
```

Invert Blurred Image then Convert to Pencil Sketch

```python
invert_blur = 255 - blurred
pencil_sketch = cv2.divide(grayscale, invert_blur, scale = 256.0)
plt.figure(figsize=(8,6))
plt.imshow(pencil_sketch,cmap='gray')
plt.axis('off')
plt.title('Pencil Sketch')
plt.show()
```
